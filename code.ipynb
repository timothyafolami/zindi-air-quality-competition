```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('Train.csv')
df_train
```

```python
df_test = pd.read_csv('Test.csv')
df_test
```

```python
submission = pd.read_csv('SampleSubmission.csv')
submission
```

```python
target = df_train['pm2_5']
# dropping the target variable from the training set
df_train = df_train.drop('pm2_5', axis=1)
```

```python
# combinig the training and test set
combined = pd.concat([df_train, df_test], axis=0).reset_index(drop=True)
combined
```

```python
# checking for total nan in it
combined.isna().sum()[:40]
```

```python
combined.isna().sum()[40:]
```

```python
# removing the id columns
data_1 = combined.drop(['id', 'site_id'], axis=1)
data_1
```

```python
# separatig the data into numerical and categorical
numerical = data_1.select_dtypes(exclude='object')
categorical = data_1.select_dtypes(include='object')
```

## Categorical Data

```python
categorical
```

```python
# working on the categorical data
# starting with the date column, extracting the year, month and day
categorical['date'] = pd.to_datetime(categorical['date'])
categorical['year'] = categorical['date'].dt.year
categorical['day'] = categorical['date'].dt.day

categorical = categorical.drop('date', axis=1)
categorical
```

```python
# checking for unique values in the categorical columns
for col in categorical.columns:
    print(col, categorical[col].nunique())
```

```python
# encoding the categorical columns
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
categorical = pd.DataFrame(encoder.fit_transform(categorical), columns=categorical.columns)
categorical
```

## Numerical data

```python
# now to the numerical data
numerical
```

```python
# filling the missing values in the numerical data by interpolating
numerical_1 = numerical.interpolate(method='linear', limit_direction='both', axis=0)
numerical_1
```

```python
# trying to remove and replace the outliers in the numerical data using quantile method
for col in numerical_1.columns:
    Q1 = numerical_1[col].quantile(0.25)
    Q3 = numerical_1[col].quantile(0.85)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    numerical_1[col] = np.where(numerical_1[col] < lower_bound, lower_bound, numerical_1[col])
    numerical_1[col] = np.where(numerical_1[col] > upper_bound, upper_bound, numerical_1[col])

numerical_1
```

```python
numerical_1['hour'].value_counts()
```

```python
numerical['hour'].value_counts()
```

```python
# checking for correlation in the numerical data with the target variable
correlation = pd.concat([numerical_1, target], axis=1).corr()
correlation['pm2_5'].sort_values(ascending=False)
```

```python
# combining the numerical and categorical data
data_2 = pd.concat([numerical_1, categorical], axis=1)
```

```python
data_2
```

```python
# scaling the data
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
data_3 = pd.DataFrame(scaler.fit_transform(data_2), columns=data_2.columns)
data_3
```

```python
# splitting the data back into training and test set
train = data_3.iloc[:len(df_train)]
test = data_3.iloc[len(df_train):].reset_index(drop=True)
```

```python
train
```

## Modeling

```python
# splitting the training set into training and validation set
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=42)
```

```python
# imporing models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
lin_pred = lin_reg.predict(x_val)
lin_rmse = np.sqrt(mean_squared_error(y_val, lin_pred))
lin_rmse
```

```python
# decision tree
tree = DecisionTreeRegressor()
tree.fit(x_train, y_train)
tree_pred = tree.predict(x_val)
tree_rmse = np.sqrt(mean_squared_error(y_val, tree_pred))
tree_rmse
```

```python
# # random forest
# forest = RandomForestRegressor(random_state=42)
# forest.fit(x_train, y_train)
# forest_pred = forest.predict(x_val)
# forest_rmse = np.sqrt(mean_squared_error(y_val, forest_pred))
# forest_rmse
```

```python
# catboost
cat = CatBoostRegressor(random_state=42, early_stopping_rounds=50, learning_rate=0.05, n_estimators=2000)
cat.fit(x_train, y_train, eval_set=(x_val, y_val))
cat_pred = cat.predict(x_val)
cat_rmse = np.sqrt(mean_squared_error(y_val, cat_pred))
cat_rmse
```

```python
# performing hyperparameter tuning on the catboost model with optuna
import optuna
def objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 100, 1000),
        'depth': trial.suggest_int('depth', 4, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.05, 0.1),
        'random_strength': trial.suggest_int('random_strength', 0, 100),
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0.0, 100.0),
        'od_type': trial.suggest_categorical('od_type', ['IncToDec', 'Iter']),
    }
    model = CatBoostRegressor(**params, random_state=42, early_stopping_rounds=50)
    model.fit(x_train, y_train, eval_set=(x_val, y_val), verbose=False)
    cat_pred = model.predict(x_val)
    return np.sqrt(mean_squared_error(y_val, cat_pred))

study = optuna.create_study(direction='minimize', study_name='catboost')
study.optimize(objective, n_trials=200, show_progress_bar=True)
```

```python
study.best_params
```

```python
# creating a new catboost model with the best parameters
cat_1 = CatBoostRegressor(**study.best_params, random_state=42, early_stopping_rounds=50)
cat_1.fit(x_train, y_train, eval_set=(x_val, y_val))
cat_1_pred = cat_1.predict(x_val)
cat_1_rmse = np.sqrt(mean_squared_error(y_val, cat_1_pred))
cat_1_rmse
```

```python
# xgboost
xgb = XGBRegressor(random_state=42, n_estimators=5100)
xgb.fit(x_train, y_train)
xgb_pred = xgb.predict(x_val)
xgb_rmse = np.sqrt(mean_squared_error(y_val, xgb_pred))
xgb_rmse
```

```python
# lightgbm
lgb = LGBMRegressor(random_state=42, n_estimators=5000)
lgb.fit(x_train, y_train)
lgb_pred = lgb.predict(x_val)
lgb_rmse = np.sqrt(mean_squared_error(y_val, lgb_pred))
lgb_rmse
```

```python
# make a prediction on the test set
test_pred = cat_1.predict(test)
test_pred
```

```python
# now adding the prediction to the submission file
submission['pm2_5'] = test_pred
```
